# Dataset Description:
Dataset describes about  state-district-wise GPDP data (GPDP:Gram Panchayat Development Plans).The dataset include 2190 rows and 10 cloums.In this dataset each state is included with 5-6 districts and the actions of GPDP are given.The dataset include 36 states. The GPDP is a structured framework aimed at improving rural development by implementing various initiatives at the Gram Panchayat level.

1.stateCode
: Unique code assigned to each state for identification.

2.stateName
: Name of the state or union territory.

3.districtCode
: Unique code assigned to each district for identification.

4.districtName
: Name of the district within the respective state.

5.finYear
: Financial year in the format 'YYYY-YYYY' (e.g., 2020-2021).

6.totalGP
: Total number of Gram Panchayats (GP) in the respective district for the given financial year.

7.sabhaHeld
: Number of sabha meetings conducted in the respective district for the given financial year.

8.gpdpApproved
: Number of GPDP (Gram Panchayat Development Plans) approved in the respective district.

9.gpdpUpload
: Number of GPDP plans uploaded online in the respective district.

10.date updated
: The date when the data was last updated, ensuring the latest available information.

# Approval rate can be calculated using gpdpapproved and total Gp.It helps in finding approval rate,that means the total plans by gp in that how many plans are approved.The lowest approval distrcit is UDALGURI,in ASSAM

```python
data["approval_rate"] = (data["gpdpApproved"] / data["totalGP"]) * 100
highest_approval_district = data.loc[data["approval_rate"].idxmax(), ["districtName","stateName", "approval_rate"]]
print(highest_approval_district)
```

# The highest approval rate district is VISHAKAPATNAM IN ANDHRA PRADESH.By this we can understand that it is having all facilities.

```python
state_gpdp = data.groupby("stateName").agg(
    total_GP=("totalGP", "sum"),
    total_approved=("gpdpApproved", "sum")
)

state_gpdp["approval_rate"] = (state_gpdp["total_approved"] / state_gpdp["total_GP"]) * 100
highest_gpdp_state = state_gpdp.loc[state_gpdp["approval_rate"].idxmax()]
lowest_gpdp_state = state_gpdp.loc[state_gpdp["approval_rate"].idxmin()]
print(highest_gpdp_state)
print(lowest_gpdp_state)
```
